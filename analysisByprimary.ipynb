```python
import numpy as np
import pylab as plb
from secbyprm import secbyprm
```

```python
data_mu = np.loadtxt("mu-by-all-prm.dat") #  By all primeries
data_mu14 = np.loadtxt("mu-by-14.dat", float) # By protons
data_mu402 = np.loadtxt("mu-by-402.dat", float) # By He
data_mu5626 = np.loadtxt("mu-by-5626.dat", float) # By Fe
```

# Muons

```python
enervals = np.array([1, 1e1, 1e2, 1e3, 1e4, 1e5], float)
minener = 0.5
maxener = 1.5

minth = 0.
maxth = 10.

mu_hist = [[], [], [], [], [], []]

read_mu = secbyprm()

hist_enermu = np.zeros([10,len(mu_hist)], float)

for i in range(9):
    hist_enermu[i] = read_mu.search(enervals, data_mu14, minener, maxener, minth, maxth, mu_hist)
    minth = maxth
    maxth += 10
```

```python
for i in range(9):
    n = (i+1)*10
    nm = i*10
    name = "By Protons "+str(nm)+"° - "+str(n)+"°"
    plb.plot(enervals, hist_enermu[i,:], 'o', markersize=18, label=name)#"By Protons < 10°")
    plb.errorbar(enervals, hist_enermu[i,:], xerr=enervals*.5, fmt='o')


#plb.errorbar(enervals, hist_ener, xerr=enervals*.5, fmt='o')

plb.legend(fontsize=24)
plb.grid()

plb.title("Count of Muons", fontsize=30)
plb.xlabel("Primary Energy / GeV", fontsize=30)
plb.ylabel("Mean of Muons / (m^2*s)", fontsize=30)

plb.xticks(fontsize=24) #htics, fontsize=20)
plb.yticks(fontsize=24) #hticsy, fontsize=20)
plb.loglog()

plb.rcParams['figure.figsize'] = 26, 14

plb.rcParams['xtick.major.size'] = 20
plb.rcParams['xtick.major.width'] = 2
plb.rcParams['xtick.minor.size'] = 10
plb.rcParams['xtick.minor.width'] = 1

plb.rcParams['ytick.major.size'] = 20
plb.rcParams['ytick.major.width'] = 2
plb.rcParams['ytick.minor.size'] = 10
plb.rcParams['ytick.minor.width'] = 1

plb.show()
```
