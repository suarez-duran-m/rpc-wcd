```python
hist_mu14 = np.zeros(int(data_mu14[:,9].max())+1, float)
for part in data_mu14[:]:
    hist_mu14[int(part[9])] += 1
    
hist_mu402 = np.zeros(int(data_mu402[:,9].max())+1, float)
for part in data_mu402[:]:
    hist_mu402[int(part[9])] += 1
    
hist_mu5626 = np.zeros(int(data_mu5626[:,9].max())+1, float)
for part in data_mu5626[:]:
    hist_mu5626[int(part[9])] += 1
```

# Neutrons

```python
data_n = np.loadtxt("n-by-all-prm.dat") #  By all primeries
data_n14 = np.loadtxt("n-by-14.dat", float) # By protons
data_n402 = np.loadtxt("n-by-402.dat", float) # By He
data_n5626 = np.loadtxt("n-by-5626.dat", float) # By Fe
```

```python
hist_n14 = np.zeros(int(data_n14[:,9].max())+1, float)
for part in data_n14[:]:
    hist_n14[int(part[9])] += 1
    
hist_n402 = np.zeros(int(data_n402[:,9].max())+1, float)
for part in data_n402[:]:
    hist_n402[int(part[9])] += 1
    
hist_n5626 = np.zeros(int(data_n5626[:,9].max())+1, float)
for part in data_n5626[:]:
    hist_n5626[int(part[9])] += 1
```

```python
plb.plot(data_n[:,0], data_n[:,1], 'or', label="By all primaries")

plb.plot(hist_n402, label="By He")
plb.plot(hist_n5626, label="By Fe")
plb.plot(hist_n14, label="By Protons")

#plb.xticks(rotation=90)
plb.legend(fontsize=24)
plb.grid()

plb.title("Count of Neutrons", fontsize=30)
plb.xlabel("Energy (E_p/GeV)", fontsize=30)
plb.ylabel("Counts / (m^2*h)", fontsize=30)

#htics = [ timedata[i] for i in range(0,len(timedata), 600) ]
#hticsy = [ i for i in range(10,28,1) ]

plb.xticks(fontsize=24) #htics, fontsize=20)
plb.yticks(fontsize=24) #hticsy, fontsize=20)
plb.loglog()
#plb.ylim(10,28)

plb.rcParams['figure.figsize'] = 26, 14

plb.show()
```

```python
mucount = 0
enervals = np.array([1, 1e1, 1e2, 1e3, 1e4, 1e5], float)
hist_ener = np.zeros(len(enervals), float)
th = 90.

shw_i = 0
tmp_index = 0

for ener in enervals:
    tmphist_mu = []    
    for line in data_n14:
        if shw_i == line[7]:
            mucount += 1
        else:
            if line[9] > 0.5*ener and line[9] < ener*1.5:
                if line[10] > 0.*th and line[10] < th:
                    shw_i = line[7]
                    tmphist_mu.append(mucount)
                    mucount = 1
    
    if len(tmphist_mu) > 0.:
        hist_mu = np.array(tmphist_mu)
        hist_ener[tmp_index] = hist_mu.mean() / 3600.
        tmp_index += 1
```

```python
plb.plot(enervals, hist_ener, 'or', label="By Protons")

plb.legend(fontsize=24)
plb.grid()

plb.title("Count of Neutrons", fontsize=30)
plb.xlabel("Primary Energy / GeV", fontsize=30)
plb.ylabel("Mean of Neutrons / (m^2*s)", fontsize=30)

plb.xticks(fontsize=24) #htics, fontsize=20)
plb.yticks(fontsize=24) #hticsy, fontsize=20)
plb.loglog()

plb.rcParams['figure.figsize'] = 26, 14

plb.rcParams['xtick.major.size'] = 20
plb.rcParams['xtick.major.width'] = 2
plb.rcParams['xtick.minor.size'] = 10
plb.rcParams['xtick.minor.width'] = 1

plb.rcParams['ytick.major.size'] = 20
plb.rcParams['ytick.major.width'] = 2
plb.rcParams['ytick.minor.size'] = 10
plb.rcParams['ytick.minor.width'] = 1

plb.show()
```
