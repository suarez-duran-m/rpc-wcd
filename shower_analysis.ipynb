```python
import numpy as np
import pylab as plb
from secbyprm import secbyprm
```

# Reading and plotting results for one hour of flux at 5100 m a.s.l.

```python
fluxground = np.loadtxt("results-3600s.hst", float) # Reading file
```

```python
plb.title("Energy Spectrum of Secondaries at 5100 m a.s.l.", fontsize=30)
plb.xlabel("Momentun / GeV", fontsize=30)
plb.ylabel("Counts / (m^2*h)", fontsize=30)

plb.plot(fluxground[:,0], fluxground[:,13], '-', linewidth=4, label="All Secondaries")
plb.plot(fluxground[:,0], fluxground[:,1], '-', linewidth=4, label="Photons")
plb.plot(fluxground[:,0], fluxground[:,2], '-', linewidth=4, label="e^-")
plb.plot(fluxground[:,0], fluxground[:,3], '-', linewidth=4, label="e^+")
plb.plot(fluxground[:,0], fluxground[:,4], '-', linewidth=4, label="mu^+")
plb.plot(fluxground[:,0], fluxground[:,5], '-', linewidth=4, label="mu^-")
plb.plot(fluxground[:,0], fluxground[:,9], '-', linewidth=4, label="n")

plb.legend(fontsize=24)
plb.grid()

plb.xticks(fontsize=24)
plb.yticks(fontsize=24)
plb.loglog()

plb.rcParams['figure.figsize'] = 26, 14

plb.rcParams['xtick.major.size'] = 20
plb.rcParams['xtick.major.width'] = 2
plb.rcParams['xtick.minor.size'] = 10
plb.rcParams['xtick.minor.width'] = 1

plb.rcParams['ytick.major.size'] = 20
plb.rcParams['ytick.major.width'] = 2
plb.rcParams['ytick.minor.size'] = 10
plb.rcParams['ytick.minor.width'] = 1

plb.show()
```

```python
data_mu = np.loadtxt("mu-by-all-prm.dat") #  By all primeries
data_mu14 = np.loadtxt("mu-by-14.dat", float) # By protons
data_mu402 = np.loadtxt("mu-by-402.dat", float) # By He
data_mu5626 = np.loadtxt("mu-by-5626.dat", float) # By Fe
```

# File Structure

# Muons

```python
hist_mu14 = np.zeros(int(data_mu14[:,9].max())+1, float)
for part in data_mu14[:]:
    hist_mu14[int(part[9])] += 1
    
hist_mu402 = np.zeros(int(data_mu402[:,9].max())+1, float)
for part in data_mu402[:]:
    hist_mu402[int(part[9])] += 1
    
hist_mu5626 = np.zeros(int(data_mu5626[:,9].max())+1, float)
for part in data_mu5626[:]:
    hist_mu5626[int(part[9])] += 1
```

```python
plb.plot(data_mu[:,0], data_mu[:,1], 'or', label="By all primaries")
plb.plot(hist_mu14, label="By Protons")
plb.plot(hist_mu402, label="By He")
plb.plot(hist_mu5626, label="By Fe")

#plb.xticks(rotation=90)
plb.legend(fontsize=24)
plb.grid()

plb.title("Count of Muons", fontsize=30)
plb.xlabel("Energy (E_p/GeV)", fontsize=30)
plb.ylabel("Counts / (m^2*h)", fontsize=30)

#htics = [ timedata[i] for i in range(0,len(timedata), 600) ]
#hticsy = [ i for i in range(10,28,1) ]

plb.xticks(fontsize=24) #htics, fontsize=20)
plb.yticks(fontsize=24) #hticsy, fontsize=20)
plb.loglog()
#plb.ylim(10,28)

plb.rcParams['figure.figsize'] = 26, 14

plb.show()
```

```python
hist_enermu14 = []
hist_enerprm = []


for part in data_mu14[:]:
    tmp = np.sqrt( part[1]*part[1] + part[2]*part[2] + part[3]*part[3] )
    hist_enermu14.append( int(tmp) )
    hist_enerprm.append( int(part[9]) )
```

```python
#plb.plot(data_p[:,0], data_p[:,1], 'or', label="By all primaries")
plb.plot(hist_enerprm, hist_enermu14, 'or', label="By Protons")
#plb.plot(hist_p402, label="By He")
#plb.plot(hist_p5626, label="By Fe")

#plb.xticks(rotation=90)
plb.legend(fontsize=24)
plb.grid()

plb.title("Count of Muons", fontsize=30)
plb.xlabel("Primary's Energy / GeV", fontsize=30)
plb.ylabel("Secundary's Energy / GeV", fontsize=30)

#htics = [ timedata[i] for i in range(0,len(timedata), 600) ]
#hticsy = [ i for i in range(10,28,1) ]

plb.xticks(fontsize=24) #htics, fontsize=20)
plb.yticks(fontsize=24) #hticsy, fontsize=20)
plb.loglog()

#plb.ylim(10,28)

plb.rcParams['figure.figsize'] = 26, 14

plb.show()
```
